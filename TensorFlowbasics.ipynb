Add and Delete Dimensions

```python
x=tf.random.uniform([28,28],maxval=10,dtype=tf.int32)
x
```

```python
x=tf.expand_dims(x,axis=2)
x
```

```python
x=tf.expand_dims(x,axis=0)
x
```

```python
x=tf.expand_dims(x,axis=-1)
x
```

when the axis of tf.expand_dims is positive it means new dimension is inserted before the current dimension and when it is negative new dimension is inserted after the current dimension

```python
import tensorflow as tf
x=tf.expand_dims(x,axis=0)
x
```

```python
x=tf.squeeze(x,axis=0)
x
```

```python
x=tf.squeeze(x)  #if axis is not mentioned ,dimension 1 is automatically deleted 
x
```

Changing the view ,adding or deleting the dimension, will not affect storage of the tensor wheras view of the tensor is changed
By swappping the dimesnion using tf.transpose() both the storage order and the view of the tensor is changed

```python
x=tf.random.normal([2,32,32,3])
x
```

```python
tf.transpose(x,perm=[0,3,1,2]) #swap dimension
```

```python
tf.transpose(x,perm=[0,2,1,3])
```
